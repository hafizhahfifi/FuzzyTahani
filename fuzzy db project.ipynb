```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt 
```

```python
x_JumlahBarang = np.arange(0, 6, 1)  
x_TotalPenjualan = np.arange(0, 2200, 1)  

JumlahBarang_sedikit = fuzz.trapmf(x_JumlahBarang,[0, 0, 1, 3])  
JumlahBarang_banyak = fuzz.trapmf(x_JumlahBarang,[2, 4, 5, 5])  

TotalPenjualan_rendah = fuzz.trapmf(x_TotalPenjualan,[0, 0, 500, 800])  
TotalPenjualan_cukup = fuzz.trimf(x_TotalPenjualan,[700, 1100, 1500])  
TotalPenjualan_tinggi = fuzz.trapmf(x_TotalPenjualan,[1500, 2000, 2200, 2200])  
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, figsize=(8, 10))

ax0.plot(x_JumlahBarang, JumlahBarang_sedikit, 'b', linewidth=1.5, label='sedikit')  
ax0.plot(x_JumlahBarang, JumlahBarang_banyak, 'g', linewidth=1.5, label='banyak')    
ax0.set_title('Jumlah Barang')
ax0.set_xlabel('Domain')
ax0.set_ylabel('Derajat Keanggotaan')
ax0.legend()  

ax1.plot(x_TotalPenjualan, TotalPenjualan_rendah, 'b', linewidth=1.5, label='rendah')  
ax1.plot(x_TotalPenjualan, TotalPenjualan_cukup, 'g', linewidth=1.5, label='cukup')  
ax1.plot(x_TotalPenjualan, TotalPenjualan_tinggi, 'r', linewidth=1.5, label='tinggi')  
ax1.set_title('Total Penjualan')
ax1.set_xlabel('Domain')
ax1.set_ylabel('Derajat Keanggotaan')
ax1.legend()  

for ax in (ax0, ax1):
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.get_xaxis().tick_bottom()
  ax.get_yaxis().tick_left()
  plt.tight_layout()
```
