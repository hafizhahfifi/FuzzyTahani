```python
import numpy as np
%pip install scikit-fuzzy
import skfuzzy as fuzz
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
x_usia = np.arange(18, 58, 1)  
x_suhubadan = np.arange(27.5, 45.0, 1)  
x_kondnapas = np.arange(14, 24, 1)  
x_jmlantbd = np.arange(0.1, 8.0, 1)
x_status = np.arange(0, 100, 1)

usia_muda = fuzz.trapmf(x_usia,[18, 18, 30, 35])  
usia_pbaya = fuzz.trapmf(x_usia,[30, 36, 45, 50])  
usia_tua = fuzz.trapmf(x_usia,[45, 50, 58, 58])  
	  
suhu_hipo = fuzz.trapmf(x_suhubadan,[27.5, 27.5, 34.5, 35.5])  
suhu_normal = fuzz.trapmf(x_suhubadan,[34.5, 35.5, 36.5, 37.5])  
suhu_hiper = fuzz.trapmf(x_suhubadan,[36.5, 37.5, 45.0, 45.0])  

kondnapas_normal = fuzz.trapmf(x_kondnapas,[14, 14, 16, 17])  
kondnapas_moderat = fuzz.trapmf(x_kondnapas,[16, 18, 19, 20])  
kondnapas_abnormal = fuzz.trapmf(x_kondnapas,[19, 20, 24, 24])  

jmlantbd_normal = fuzz.trapmf(x_jmlantbd,[0.1, 0.1, 1.5, 2.0])  
jmlantbd_moderat = fuzz.trapmf(x_jmlantbd,[1.5, 2.1, 3.8, 4.0])  
jmlantbd_abnormal = fuzz.trapmf(x_jmlantbd,[3.8, 4.0, 8.0, 8.0])

status_negatif = fuzz.trapmf(x_status,[0, 0, 49, 50])
status_positif = fuzz.trapmf(x_status,[50, 51, 100, 100])
```

```python
fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(nrows=5, figsize=(8, 10))

ax0.plot(x_usia, usia_muda, 'b', linewidth=1.5, label='Muda')  
ax0.plot(x_usia, usia_pbaya, 'g', linewidth=1.5, label='Paruhbaya')  
ax0.plot(x_usia, usia_tua, 'r', linewidth=1.5, label='Tua')  
ax0.set_title('Indeks Usia Pasien')  
ax0.legend()  
	  
ax1.plot(x_suhubadan, suhu_hipo, 'b', linewidth=1.5, label='Hipotermia')  
ax1.plot(x_suhubadan, suhu_normal, 'g', linewidth=1.5, label='Normal')  
ax1.plot(x_suhubadan, suhu_hiper, 'r', linewidth=1.5, label='Hipertermia')  
ax1.set_title('Indeks Suhu Badan')  
ax1.legend()  
	  
ax2.plot(x_kondnapas, kondnapas_normal, 'b', linewidth=1.5, label='Normal')  
ax2.plot(x_kondnapas, kondnapas_moderat, 'g', linewidth=1.5, label='Moderat')  
ax2.plot(x_kondnapas, kondnapas_abnormal, 'r', linewidth=1.5, label='Abnormal')  
ax2.set_title('Indeks Kondisi Pernapasan')  
ax2.legend()

ax3.plot(x_jmlantbd, jmlantbd_normal, 'b', linewidth=1.5, label='Normal')  
ax3.plot(x_jmlantbd, jmlantbd_moderat, 'g', linewidth=1.5, label='Moderat')  
ax3.plot(x_jmlantbd, jmlantbd_abnormal, 'r', linewidth=1.5, label='Abnormal')  
ax3.set_title('Indeks Jumlah Antibodi')  
ax3.legend()

ax4.plot(x_status, status_negatif, 'g', linewidth=1.5, label='Negatif')
ax4.plot(x_status, status_positif, 'r', linewidth=1.5, label='Positif')
ax4.set_title('Indeks Status Pasien')  
ax4.legend()

for ax in (ax0, ax1, ax2, ax3, ax4):
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.get_xaxis().tick_bottom()
  ax.get_yaxis().tick_left()
  plt.tight_layout()
```
